```python
import gurobipy as gp
from gurobipy import GRB

model = gp.Model("hw0")
```

```python
model.printStats()
```

```python
# create variables 
x1= model.addVar(vtype=GRB.BINARY, name="x1")
x2 = model.addVar(vtype=GRB.BINARY, name="x2")

model.setObjective(3*x1 + 2*x2, GRB.MAXIMIZE)

model.addConstr(2*x1 + 3*x2 <= 4, "contraint1")
```

```python
model.printStats()
```

```python
model.optimize()
```
